```python
# Counting all the unique values in '[substans type(gift#)]'
import pandas as pd

def count_unique_substans_types(df, prefix=""):
    all_types = []
    for i in range(1, 16):
        col_subst = f'[substans type(gift{i})]'.lower()
        if col_subst in df.columns:
            # Drop NAs
            vals = df[col_subst].dropna().tolist()
            all_types.extend(vals)
    # Now count unique values
    type_counts = pd.Series(all_types).value_counts()

    print(f"\n=== {prefix} Unique values in all [substans type(gift#)] columns, counted: ===\n")
    print(type_counts)

# For both sheets:
count_unique_substans_types(young_filtered, "Alder<70")
count_unique_substans_types(old_filtered, "Alder>70")
```

```python
# BIO elimination
import pandas as pd

def combined_bio_name_counts(df, prefix=""):
    # To collect all names (both generisk and navn) when type is BIO
    all_names = []

    for i in range(1, 16):
        col_subst = f'[substans type(gift{i})]'.lower()
        col_gn = f'[generisk navn(gift{i})]'.lower()
        col_navn = f'[navn(gift{i})]'.lower()

        # Ensure columns exist
        if all(col in df.columns for col in [col_subst, col_gn, col_navn]):
            mask = df[col_subst] == 'BIO'

            generics = df.loc[mask, col_gn].dropna().tolist()
            navns = df.loc[mask, col_navn].dropna().tolist()

            all_names.extend(generics)
            all_names.extend(navns)

    # Count all unique names (case-insensitive, optional)
    name_counts = pd.Series(all_names).value_counts()

    print(f"\n=== {prefix} Combined unique values ([generisk navn] and [navn]), counted: ===\n")
    print(name_counts)

# For both sheets:
combined_bio_name_counts(young_filtered, "Alder<70")
combined_bio_name_counts(old_filtered, "Alder>70")
```

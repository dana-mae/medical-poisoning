```python
#Finding unique values and counting them
import pandas as pd

def get_unique_gift_names(df, max_gifts=15):
    generisk_cols = [f'[generisk navn(gift{i})]'.lower() for i in range(1, max_gifts+1)]
    navn_cols = [f'[navn(gift{i})]'.lower() for i in range(1, max_gifts+1)]
    # Combine both lists
    gift_cols = generisk_cols + navn_cols
    # Only keep columns that are actually present (defensive)
    gift_cols = [col for col in gift_cols if col in df.columns]
    # Stack all values into one Series
    all_values = pd.Series(dtype=str)
    for col in gift_cols:
        all_values = pd.concat([all_values, df[col].dropna().astype(str).str.lower()])
    # Get value counts
    value_counts = all_values.value_counts().sort_values(ascending=False)
    return value_counts

for group_name, group_df in zip(['young', 'old'], [young, old]):
    print(f"Unique '[generisk navn]' and '[navn]' in {group_name}:")
    counts = get_unique_gift_names(group_df, max_gifts=15)
    print(counts)
    print()
```

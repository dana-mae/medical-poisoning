```python
#Finding unique values and counting them
import numpy as np
from collections import Counter

# Define the columns
generisk_cols = [f'[generisk navn(gift{i})]' for i in range(1, 16)]
navn_cols = [f'[navn(gift{i})]' for i in range(1, 16)]

# All relevant columns (already lowercased in your code)
columns = [c.lower() for c in generisk_cols + navn_cols]

# Extract the values from both dataframes
def get_values(df, columns):
    vals = []
    for col in columns:
        if col in df.columns:
            # dropna (empty cells)
            vals.extend(df[col].dropna().astype(str))
    return vals

all_vals = get_values(young, columns) + get_values(old, columns)

# Normalize and clean: lowercase, strip whitespace, remove empty/nan/none
cleaned_vals = []
for val in all_vals:
    v = val.lower().strip()
    if not v or v in ['nan', 'none', '']:  # skip empty/nan entries
        continue
    v = v.replace(',', '')  # Remove commas so output is comma-free
    cleaned_vals.append(v)

# Count unique values
counter = Counter(cleaned_vals)

# Write output: name TAB count, one per line
with open('raw_unique_values.csv', 'w', encoding='utf-8') as f:
    for name, count in counter.most_common():
        f.write(f"{name}\t{count}\n")

print(f"Done. Found {len(counter)} unique names. Output is in 'raw_unique_values.csv'.")
```

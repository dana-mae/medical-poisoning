```python
"show our unique values. use 'anden årsag' for other values"
import numpy as np

allowed_values = [
    'ingen risiko',
    'risiko for begrænset forgiftning',
    'risiko for manifest forgiftning',
    np.nan,  # for missing entries
    'uafklaret risiko',
    'risiko for livstruende forgiftning'
]

def map_risiko(val):
    # consider nan as allowed
    if pd.isna(val):
        return np.nan
    # strip whitespace just in case
    val_stripped = str(val).strip()
    if val_stripped in allowed_values:
        return val_stripped
    else:
        return 'anden årsag'

young['risiko'] = young['[ingen risiko(risiko)]'].apply(map_risiko)
old['risiko'] = old['[ingen risiko(risiko)]'].apply(map_risiko)

# show what the new column looks like
print("value counts for 'risiko' - young:")
print(young['risiko'].value_counts(dropna=False))
print()
print("value counts for 'risiko' - old:")
print(old['risiko'].value_counts(dropna=False))
```

```python
#removing rows
import numpy as np

# columns to check if they are empty
empty_gift1_cols = [
    '[generisk navn(gift1)]',
    '[navn(gift1)]',
    '[substans type(gift1)]'
]

def filter_dataframe(df):
    """
    remove rows where:
      1. all of '[generisk navn(gift1)]', '[navn(gift1)]', and '[substans type(gift1)]' are empty/nan/''
      2. and '[ingen risiko(risiko)]' is either 'ingen risiko' or is nan
      3. and '[akut forgiftning(forespørgselsart)]' is nan
    """
    # 1. are all three gift1 columns missing or empty?
    gift1_all_empty = df[empty_gift1_cols].replace('', np.nan).isna().all(axis=1)

    # 2. is '[ingen risiko(risiko)]' == 'ingen risiko' or nan?
    ingen_risiko_col = '[ingen risiko(risiko)]'
    ingen_risiko_is_bad = (df[ingen_risiko_col].isna()) | (df[ingen_risiko_col] == 'ingen risiko')

    # 3. is '[akut forgiftning(forespørgselsart)]' nan?
    akut_forgiftning_col = '[akut forgiftning(forespørgselsart)]'
    akut_is_nan = df[akut_forgiftning_col].isna()

    # find rows where all conditions are true
    rows_to_remove = gift1_all_empty & ingen_risiko_is_bad & akut_is_nan

    # invert the mask to keep only rows not matching all conditions
    filtered_df = df[~rows_to_remove].copy()

    # optionally, print how many rows were removed
    print(f"rows removed: {rows_to_remove.sum()}")
    print(f"rows remaining: {filtered_df.shape[0]}")
    return filtered_df

# apply to both dataframes
young_filtered = filter_dataframe(young)
old_filtered = filter_dataframe(old)

# if you want to replace the originals:
# young = young_filtered
# old = old_filtered
```

```python
"searching for unique values, so we can search only for medicines"
import pandas as pd

# list of columns to inspect
substans_cols = [
    '[substans type(gift1)]',
    '[substans type(gift2)]',
    '[substans type(gift3)]',
    '[substans type(gift4)]',
    '[substans type(gift5)]',
    '[substans type(gift6)]',
    '[substans type(gift7)]',
    '[substans type(gift8)]',
    '[substans type(gift9)]',
    '[substans type(gift10)]',
    '[substans type(gift11)]',
    '[substans type(gift12)]',
    '[substans type(gift13)]',
    '[substans type(gift14)]',
    '[substans type(gift15)]',
]

# combine all values from the columns, filter nan and blanks, and get unique ones
def get_unique_substans_types(df):
    # concatenate all values from the columns into a single series
    all_values = pd.Series(dtype="object")
    for col in substans_cols:
        all_values = pd.concat([all_values, df[col]], ignore_index=True)
    # remove nan and blank values
    all_values = all_values.dropna()
    all_values = all_values[all_values.astype(str).str.strip() != '']
    # get unique values and sort them
    unique_values = sorted(all_values.unique())
    return unique_values

# example for young:
unique_substans_types_young = get_unique_substans_types(young_filtered)
print("unique substans types in young_filtered:")
print(unique_substans_types_young)

# and for old:
unique_substans_types_old = get_unique_substans_types(old_filtered)
print("\nunique substans types in old_filtered:")
print(unique_substans_types_old)
```

```python
# Count all unique substans type
# List of substance types to count
wanted_types = ['AMED', 'BIO', 'DIVERSE', 'KEMI', 'MED', 'RUS', 'SPECIALDONOTDELETE', 'ZZMED']

substans_cols = [
    '[substans type(gift1)]',
    '[substans type(gift2)]',
    '[substans type(gift3)]',
    '[substans type(gift4)]',
    '[substans type(gift5)]',
    '[substans type(gift6)]',
    '[substans type(gift7)]',
    '[substans type(gift8)]',
    '[substans type(gift9)]',
    '[substans type(gift10)]',
    '[substans type(gift11)]',
    '[substans type(gift12)]',
    '[substans type(gift13)]',
    '[substans type(gift14)]',
    '[substans type(gift15)]',
]

def count_substance_types(df):
    # Stack all values into one Series
    all_types = df[substans_cols].values.ravel()
    # Convert to Series, drop NaN and remove blanks
    all_types = pd.Series(all_types)
    all_types = all_types.dropna()
    all_types = all_types[all_types.astype(str).str.strip() != '']
    # Count only the wanted types
    counts = all_types.value_counts()
    return {typ: counts.get(typ, 0) for typ in wanted_types}

young_counts = count_substance_types(young_filtered)
old_counts = count_substance_types(old_filtered)

print("Counts in young_filtered:")
for typ, cnt in young_counts.items():
    print(f"{typ}: {cnt}")

print("\nCounts in old_filtered:")
for typ, cnt in old_counts.items():
    print(f"{typ}: {cnt}")
```

```python
#REsearching substans type "SPECIALDONOTDELETE"
from collections import Counter, defaultdict

def count_navn_specialdonotdelete(df):
    navn_vals = []
    display_dict = defaultdict(Counter)  # to track variants for display
    for i in range(1, 16):
        substans_col = f'[substans type(gift{i})]'
        navn_col = f'[navn(gift{i})]'
        # Select where substans type (case-insensitive) is 'SPECIALDONOTDELETE'
        mask = df[substans_col].astype(str).str.lower() == 'specialdonotdelete'
        names = df.loc[mask, navn_col].dropna().astype(str).str.strip()
        for v in names:
            v_lower = v.lower()
            navn_vals.append(v_lower)
            display_dict[v_lower][v] += 1
    # Count each unique value (case-insensitive)
    counts = Counter(navn_vals)
    # To display, use the most common actual form (casing) found in the data
    display_name = {k: max(variants, key=variants.get) 
                    for k, variants in display_dict.items()}
    # Create a sorted list of tuple pairs (display name, count)
    sorted_counts = sorted(
        [(display_name[k], v) for k, v in counts.items()],
        key=lambda x: (-x[1], x[0].lower())
    )
    return sorted_counts

# For young_filtered
special_young_counts = count_navn_specialdonotdelete(young_filtered)
print("Counts for [navn(gift#)] where substans type is 'SPECIALDONOTDELETE' (young_filtered):")
for navn, count in special_young_counts:
    print(f"{navn}: {count}")

# For old_filtered
special_old_counts = count_navn_specialdonotdelete(old_filtered)
print("\nCounts for [navn(gift#)] where substans type is 'SPECIALDONOTDELETE' (old_filtered):")
for navn, count in special_old_counts:
    print(f"{navn}: {count}")
```

```python
"Removing SPECIALDONOTDELETE rows that are not medicine, changing the substans type to MED"
import pandas as pd
from IPython.display import display  # use display() for better formatting in notebooks

# Prepare your medicine set for fast lookup
medicine_set = set([m.lower().strip() for m in known_medicines])

def clean_specialdonotdelete_rows(df, medicine_set, show_examples=5):
    df = df.copy()  # avoid mutating the input frame
    deleted_indices = set()

    # Track which rows have had 'SPECIALDONOTDELETE' but not matched a medicine
    for i in range(1, 16):
        subst_col = f'[substans type(gift{i})]'
        navn_col = f'[navn(gift{i})]'

        mask_special = df[subst_col].astype(str).str.lower() == 'specialdonotdelete'
        navn_vals = df.loc[mask_special, navn_col].astype(str).str.strip().str.lower()

        # Where navn is a recognized medicine
        is_medicine = navn_vals.isin(medicine_set)
        idx_special = df.loc[mask_special].index

        # 1. Change to MED where navn is in medicine_set
        med_indices = idx_special[is_medicine]
        df.loc[med_indices, subst_col] = 'MED'

        # 2. Else: collect indices for deletion
        not_med_indices = idx_special[~is_medicine]
        deleted_indices.update(not_med_indices)

    # Final result: drop all those rows (if any)
    if deleted_indices:
        deleted_indices = list(sorted(deleted_indices))
        deleted_df = df.loc[deleted_indices].copy()
        remaining_df = df.drop(deleted_indices).copy()
    else:
        deleted_df = pd.DataFrame(columns=df.columns)
        remaining_df = df.copy()

    # Report
    print(f"Rows deleted: {len(deleted_df)}")
    if len(deleted_df):
        print("Examples of deleted rows:")
        display(deleted_df.head(show_examples))
    else:
        print("No rows deleted.")

    return remaining_df, deleted_df

# Usage:
print("Processing young_filtered:")
young_cleaned, young_deleted = clean_specialdonotdelete_rows(young_filtered, medicine_set)

print("\nProcessing old_filtered:")
old_cleaned, old_deleted = clean_specialdonotdelete_rows(old_filtered, medicine_set)

# Results in: young_cleaned, old_cleaned as your new processed DataFrames
# and young_deleted, old_deleted as your deletions, with display output
```

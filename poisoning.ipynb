```python
import pandas as pd

cause_cols = [
    '[Ulykke/uheld(Årsag)]',
    '[Leg(Årsag)]',
    '[Suicidal/Affekt(Årsag)]',
    '[Misbrug(Årsag)]',
    '[Forveksling(Årsag)]',
    '[Omhældning(Årsag)]',
    '[Fejldosering(Årsag)]',
    '[Levnedsmiddel(Årsag)]',
]

young['Cause'] = young[cause_cols].agg(lambda row: ', '.join([str(v) for v in row if pd.notnull(v) and str(v).strip()]), axis=1)
# Shows only rows where 'Cause' contains at least one comma+space
print(young[young['Cause'].str.contains(', ', na=False)][['Cause']].head(100))

old['Cause'] = old[cause_cols].agg(lambda row: ', '.join([str(v) for v in row if pd.notnull(v) and str(v).strip()]), axis=1)
# Shows only rows where 'Cause' contains at least one comma+space
print(old[old['Cause'].str.contains(', ', na=False)][['Cause']].head(100))
```

```python
# Helper function: extracts unique causes from Cause column
def get_unique_causes(df):
    return set(
        cause.strip()
        for entry in df['Cause'].dropna()
        for cause in entry.split(",")
        if cause.strip()
    )

young_unique_causes = get_unique_causes(young)
old_unique_causes = get_unique_causes(old)

print("Unique causes in young:")
print(young_unique_causes)

print("\nUnique causes in old:")
print(old_unique_causes)
```

```python
import numpy as np

selected_causes = {
    'Suicidal/Affekt',
    'Forveksling',
    'Omhældning',
    'Levnedsmiddel',
    'Leg',
    'Misbrug',
    'Ulykke/uheld',
    'Fejldosering'
}

def filter_causes(cause_entry, selected_causes):
    if pd.isnull(cause_entry):
        return None
    filtered = [c.strip() for c in cause_entry.split(',') if c.strip() in selected_causes]
    return ', '.join(filtered) if filtered else None

def classify_rows(df, dataframe_name="Data"):
    total_rows = len(df)
    # Identify empty cells
    empty_mask = df['Cause'].isnull() | (df['Cause'].str.strip() == "")
    num_empty = empty_mask.sum()
    
    # Rows that are NOT empty originally
    not_empty_mask = ~empty_mask
    # Store original (before filtering)
    original_cause = df.loc[not_empty_mask, 'Cause']

    # Filter for selected causes
    filtered_cause = original_cause.apply(lambda x: filter_causes(x, selected_causes))
    # "bad data" = rows that WERE not empty, but BECOME empty after filtering
    bad_data_mask = filtered_cause.isnull() | (filtered_cause.str.strip() == "")
    num_bad = bad_data_mask.sum()

    # "good" data is everything else
    num_good = total_rows - num_empty - num_bad

    # Percentages
    print(f"===== {dataframe_name} =====")
    print(f"Total rows:       {total_rows}")
    print(f"Empty:            {num_empty} ({num_empty/total_rows:.1%})")
    print(f"Bad data:         {num_bad} ({num_bad/total_rows:.1%})")
    print(f"Valid after filter:{num_good} ({num_good/total_rows:.1%})")
    print()

    # --- If you want to actually produce the cleaned dataframe:
    #    here we update filtered causes, but only for rows originally non-empty!
    df_filtered = df.copy()
    df_filtered.loc[not_empty_mask, 'Cause'] = filtered_cause
    # Keep only rows that have valid Cause after filtering
    df_valid = df_filtered[df_filtered['Cause'].notnull() & (df_filtered['Cause'].str.strip() != "")]
    return df_valid

# Apply for young and old
young_valid = classify_rows(young, "Young")
old_valid = classify_rows(old, "Old")

# Now young_valid and old_valid only have "good" rows (the rest are classified above)
```
